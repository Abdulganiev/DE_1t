```python
import pandas as pd
```

```python
df = pd.read_csv('ds_salaries.csv')
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
df.head()
```

### Сгруппируйте вакансии по направлениями (DS, DE, Software Engenering, etc.)

```python
def replace_vacan(v):
    vacan = {
        'DS' : ['Scientist', 'Science', 'Architect', 'Machine Learning', 'Data Specialist', 'Head of Data'], 
        'DE' : ['Engineer'], 
        'DA' : ['Analyst', 'Analytics'], 
        'Software Engenering' : ['Developer'],
        'etc': ['3D'],
    }
    for key, values in vacan.items():
        for value in values:
            if value in v:
                return key
```

```python
df['area_of_activity'] = df['job_title'].apply(replace_vacan)
```

```python
df['area_of_activity'].value_counts()
```

### Какая средняя и медианная зарплата по группам вакансий?

```python
df.groupby(['area_of_activity']).agg({'salary_in_usd' : ['mean', 'median']})
```

### Какая средняя и медианная зарплата по каждому региону?

```python
df.groupby(['company_location']).agg({'salary_in_usd' : ['mean', 'median']})
```

### Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?

```python
pd.DataFrame(df.groupby(['area_of_activity'])['salary_in_usd'].mean()).sort_values('salary_in_usd', ascending=False).head(1)
```

### Какое процентное соотношение каждого региона по вакансиям от всех вакансий?

```python
all_vacan = df['company_location'].count()
for el in pd.DataFrame(df.groupby(['company_location'])['job_title'].count()).itertuples():
    vacan_ = (el[1] / all_vacan) * 100
    print(el[0], ' - ' , el[1], ' - ' , round(vacan_, 2), '%', sep='')
```

### Какая корреляция уровня опыта от зарплаты?

```python
def level(a):
    experience_level = {
        'EN' : 1, 
        'EX' : 2, 
        'MI' : 3, 
        'SE' : 4
    }
    return experience_level[a]
```

```python
df['experience_level_code'] = df['experience_level'].apply(level)
```

```python
print('корреляция уровня опыта от зарплаты - ', df['experience_level_code'].corr(df['salary_in_usd']))
```
