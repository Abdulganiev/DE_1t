```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AffinityPropagation
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
import plotly as py
import plotly.graph_objs as go
import os
py.offline.init_notebook_mode(connected = True)
#print(os.listdir("../input"))
import datetime as dt
import missingno as msno
plt.rcParams['figure.dpi'] = 140

color_rectangle = ['#000000', '#FFFFFF', '#D6001C', '#F5F5F5', '#D7D7D7', '#6A737B']
```

```python
df = pd.read_csv('DataEngineer.csv')
```

#### Построить график компаний с самым большим кол-вом вакансий

```python
company = df['Company Name'].value_counts().head(10)

fig, ax = plt.subplots(1, 1, figsize=(20, 8))
ax.bar(company.index, company, width=0.6, color=color_rectangle, alpha=0.9, label='company',
       edgecolor = 'darkblue')
for i in company.index:
    ax.annotate(f"{company[i]}", 
                   xy = (i, company[i]+0.5)
                , fontsize=15
                , va = 'center'
                , ha='center'
                , fontweight='light'
                , fontfamily='serif'
                , 
               )
ax.legend().set_visible(False)
fig.text(0.2, 0.9, 'График компаний с самым большим кол-вом вакансий', fontsize=20, fontweight='bold', fontfamily='serif')
plt.show()
```

#### Построить график самых популярный названий в вакансиях

```python
Job = df['Job Title'].value_counts().head(10)

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
ax.pie(Job, labels=Job.index, autopct= '%0.1f%%', colors=color_rectangle, 
       pctdistance=1.1, labeldistance=1.2, wedgeprops={'lw':1, 'ls':'--','edgecolor':"k"})
fig.text(0.2, 0.9, 'График самых популярных вакансий', fontsize=15, fontweight='bold', fontfamily='serif')
plt.show()
```

#### Построить график самых популярных местоположений вакансий по их численности и построить круговую диаграмму

```python
location = df['Location'].value_counts().head(10)

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
ax.pie(location, labels=location.index, autopct= '%0.1f%%', colors=color_rectangle, 
       pctdistance=1.1, labeldistance=1.2, wedgeprops={'lw':1, 'ls':'--','edgecolor':"k"})
fig.text(0.2, 0.9, 'График самых популярных местоположений вакансий по их численности', 
         fontsize=15, fontweight='bold', fontfamily='serif')
plt.show()
```

#### Построить график популярности по секторам работы

```python
sector = df[df['Sector'] != '-1']['Sector'].value_counts().head(10)
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
ax.pie(sector, labels=sector.index, autopct='%0.1f%%', colors=color_rectangle, 
       pctdistance=1.1, labeldistance=1.2, wedgeprops={'lw':1, 'ls':'--','edgecolor':"k"})
fig.text(0.2, 0.9, 'График популярности по секторам работы', fontsize=15, fontweight='bold', fontfamily='serif')
plt.show()
```

#### Построить график используя сектора работы и средний уровень зарплат в них

```python
def df_split(x):
    try:
        x = x.replace('(Glassdoor est.)', '').replace('(Employer est.)', '').replace('$', '').replace(' ', '')
        a, b = x.split('-')
        a = int(a.replace('K', ''))
        b = int(b.replace('K', ''))
        return round((a+b)/2, 2)
    except:
        return None
```

```python
df['Salary mean'] = df['Salary Estimate'].apply(df_split)
```

```python
sector_salary = df[df['Sector'] != '-1'].groupby('Sector')['Salary mean'].mean()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.barh(sector_salary.index, sector_salary
        , color=color_rectangle, alpha=0.9
        , label='sector_salary', edgecolor = 'darkblue')
for i in sector_salary.index:
    ax.annotate(f"${round(sector_salary[i], 2)}K", 
                   xy = (sector_salary[i]+10, i)
                , fontsize=15
                , va = 'center'
                , ha='center'
                , fontweight='light'
                , fontfamily='serif'
                , 
               )
ax.legend().set_visible(False)
fig.text(0.2, 0.9, 'График сектора работы и средний уровень зарплат в них', 
         fontsize=20, fontweight='bold', fontfamily='serif')
plt.show()
```

#### Построить описаний вакансий облако тегов, исследовать самые популярные слова

```python
from wordcloud import WordCloud
import matplotlib
import matplotlib.pylab as plt

vacancy_names = df['Job Title']

cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", color_rectangle)

text = str(list(df['Job Title'])).replace(',', '').replace('[', '').replace("'", '').replace(']', '').replace('.', '')

wordcloud = WordCloud(width = 500, height = 200, max_words = 150
                      , min_font_size = 8
                      , background_color = 'white'
                      , colormap=cmap
                     ).generate(text)


plt.figure(figsize = (16, 16)) 
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.imshow(wordcloud)
```
