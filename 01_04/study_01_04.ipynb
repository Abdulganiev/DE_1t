```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 10) # y = x
y = np.arange(-10, 10)**2 + 3 # y = x^2 + k
```

```python
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
import time
from IPython import display
```

```python
for k in range(-50, 50):
  x = np.arange(-10, 10)
  y = np.arange(-10, 10) * k
  plt.plot(x, y)
  plt.grid()
  display.display(plt.gcf())
  display.clear_output(wait=True)
  time.sleep(1)
```

### Датасет  
возраст;пол;индекс веса;кол-во детей;курит да/нет;регион;затраты на страховку.  
**Задача** - прогнозирование величины страховки, т.е.  
y - затраты на страховку  
x - возраст;пол;индекс веса;кол-во детей;курит да/нет;регион

```python
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charhes = []

with open('task 01_04_dataset_home.txt', 'r')  as file:
  for line in file.readlines():
    age.append(int(line.split(',')[0]))

    if line.split(',')[1] == 'male':
      sex.append(1)
    else:
      sex.append(0)

    bmi.append(float(line.split(',')[2]))

    children.append(int(line.split(',')[3]))

    if line.split(',')[4] == 'yes':
      smoker.append(1)
    else:
      smoker.append(0)

    if   line.split(',')[5] == 'northeast':
      region.append(0)
    elif line.split(',')[5] == 'northwest':
      region.append(1)
    elif line.split(',')[5] == 'southeast':
      region.append(2)
    elif line.split(',')[5] == 'southwest':
      region.append(3)
    else:
      region.append(line.split(',')[5])

    charhes.append(float(line.split(',')[6]))
```

```python
age = np.array(age)
sex = np.array(sex)
bmi = np.array(bmi)
children = np.array(children)
smoker = np.array(smoker)
region = np.array(region)
charhes = np.array(charhes)
```
