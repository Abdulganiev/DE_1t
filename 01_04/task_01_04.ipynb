### Задание 1. Построение распределений  

Сгенерировать выборку с распределением Рэлэя ( https://ru.wikipedia.org/wiki/Распределение_Рэлея ) используя numpy и scipy и построить на основе ее кривую плотности распределения на основе гистограммы из **Matplotlib**.

```python
import numpy as np
import matplotlib.pyplot as plt # визуализация данных
from IPython import display
import time
```

```python
x = np.arange(0.1, 10, 0.001) 
plt.grid()
for q in range(1, 6):
    y = (x / q**2) * np.exp(-(x**2 / (2 * q**2)))
    plt.plot(x, y, label = f'q={q}')
    print(q)
    plt.legend()
    display.display(plt.gcf())
    display.clear_output(wait=True)
    time.sleep(1)
```

Для анализа используйте графики **scatter**, **plot**, **hist**, **bar** из **matplotlib** и функции **numpy** и **scipy**.

возраст;пол;индекс веса;кол-во детей;курит да/нет;регион;затраты на страховку.  
Задача - прогнозирование величины страховки, т.е.

[dataset_home](./task_01_04_dataset_home.txt)  

```python
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charhes = []

with open('task_01_04_dataset_home.txt', 'r')  as file:
    for line in file.readlines():
        age.append(int(line.split(',')[0]))

        if line.split(',')[1] == 'male':
            sex.append(1)
        else:
            sex.append(0)
    
        bmi.append(float(line.split(',')[2]))
    
        children.append(int(line.split(',')[3]))
    
        if line.split(',')[4] == 'yes':
            smoker.append(1)
        else:
            smoker.append(0)
    
        if   line.split(',')[5] == 'northeast':
            region.append(0)
        elif line.split(',')[5] == 'northwest':
            region.append(1)
        elif line.split(',')[5] == 'southeast':
            region.append(2)
        elif line.split(',')[5] == 'southwest':
            region.append(3)
        else:
            region.append(line.split(',')[5])
    
        charhes.append(float(line.split(',')[6]))
```

```python
age = np.array(age)
sex = np.array(sex)
bmi = np.array(bmi)
children = np.array(children)
smoker = np.array(smoker)
region = np.array(region)
charhes = np.array(charhes)
```

```python
print('Данные по респондентам:')
print(f'Возраст: максимальный - {age.max()}, минимальный - {age.min()}, средний - {round(age.mean())}')
plt.hist(age)
plt.grid()
plt.show()
print(f'Индекс веса: максимальный - {bmi.max()}, минимальный - {bmi.min()}, средний - {round(bmi.mean(), 3)}')
plt.hist(bmi)
plt.grid()
plt.show()
print(f'Всего - {sex.shape[0]}, из них мужчин - {sex.sum()}, женщин - {sex[sex<1].shape[0]}.')
plt.hist(sex)
plt.grid()
plt.show()
print(f'Курящих - {smoker[smoker>0].shape[0]}, некурящих - {smoker[smoker<1].shape[0]}.')
plt.hist(smoker)
plt.grid()
plt.show()
print(f'Расходы на страховку: максимальный - {charhes.max()}, минимальный - {charhes.min()}, средний - {round(charhes.mean(), 4)}')
plt.hist(charhes, orientation='horizontal')
plt.grid()
plt.show()
```
